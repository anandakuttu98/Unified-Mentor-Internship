# Data Analyst Jobs

## Introduction

The Pandemic caused quite a stir in the job market and especially in the field of data analytics. The dataset posted in the website SimplyHired could be very useful in performing wide-ranging analysis and data visualisation so as to infer valuable insights which can assist in smooth and efficient job search. This dataset was created by picklesuet and contains more than 2,000 job listings for data analyst positions.

## Dataset Overview

Before digging deeper into the data for data analysis purposes, we need to get an overview of the different columns and variables of the dataset.

The dataset consists of the following information: -
1. **Job Title**: - A name that describes someone's job or position.
2. **Salary Estimate**: A display of a range for annual base or hourly pay and are specific to Data Analytics Industry.
3. **Job Description**: - The plain-language tool that explains the tasks, duties, function and responsibilites of a position.
4. **Rating**: - Company Rating
5. **Company Name**: - The name of the company.
6. **Location**: - The location where the job is available.
7. **Headquarters**: - The headquarters of the company.
8. **Size**: - The size of the employees
9. **Type of Ownership**: - Type of ownership whether it is public or private or non-profit.
10. **Industry**: - Different industries where the job is available.
11. **Sector**: - Sector where the job is available.
12. **Revenue**: - Company earnings annually.
13. **Easy Apply**: - Easy Apply section.

Observation: -
There are 2253 rows and 13 columns and 1 missing values.

### Problem Statement

To analyse the trend of data analyst job postings based on factors like salary, ratings, size, revenue and so on.

### Objectives

1. To understand different factors in the dataset.
2. To graphically analyse the factors affecting the job postings.
3. To draw inference from the trends in the dataset.

## I. Data Collection

```python
# 1. Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import warnings
warnings.filterwarnings('ignore')
```

```python
# 2. Importing the Dataset
data_analyst_jobs = pd.read_csv('/content/DataAnalyst.csv', index_col = 0)
data_analyst_jobs.head()
```

## 2. Data Cleaning

```python
# 3. Cleaning the data
data_analyst_jobs.drop(columns = ['Founded', 'Competitors'], axis = 1, inplace= True)
data_analyst_jobs.head()
```

```python
# 4. Describe the dataset
data_analyst_jobs.info()
```

```python
print(f'Number of rows: {len(data_analyst_jobs.index)}; Number of columns: {len(data_analyst_jobs.columns)}; Number of missing values: {sum(data_analyst_jobs.isna().sum())}')
```

```python
# 5. Changing the name of the columns
data_analyst_jobs.rename(columns= {'Job Title': 'job_title',
                                   'Salary Estimate': 'salary_estimate',
                                   'Job Description': 'job_description',
                                   'Rating': 'rating',
                                   'Company Name': 'company_name',
                                   'Location': 'location',
                                   'Headquarters': 'headquarters',
                                    'Size': 'size',
                                  'Type of ownership': 'type_of_ownership',
                                  'Industry': 'industry',
                                  'Sector': 'sector',
                                  'Revenue': 'revenue',
                                  'Easy Apply': 'easy_apply'
                                  }, inplace = True)
```

```python
# 6. Replacing the duplicate values in the Job title Column
data_analyst_jobs['job_title'].replace(['Sr. Data Analyst',
                                        'sr. data analyst',
                                        'Sr Data Analyst',
                                        'sr data analyst',
                                        'senior data analyst',
                                        'Senior Data Analyst',
                                        'Data Analyst III',
                                        'data analyst iii',
                                        'senior data analyst'],
                                       'Senior Data Analyst', regex=True, inplace = True )
```

```python
data_analyst_jobs['job_title'].replace(['Data Analyst I',
                                        'data analyst i',
                                        'Data Analyst Junior',
                                        'data analyst junior',
                                        'Junior Data Analyst',
                                        'Junior Data AnalystI',
                                        'Junior Data Analystl'],
                                       'Junior Data Analyst', regex=True, inplace = True)
```

```python
data_analyst_jobs['job_title'].replace(['Data Analyst II',
                                        'data analyst ii',
                                        'Middle Data Analyst'],
                                       'Middle Data Analyst', regex=True, inplace = True)
```

```python
data_analyst_jobs.head()
```

# **3. Salary Estimates and other Factors**

```python
# 7. To find the average salary
data_analyst_jobs[['MinSalary', 'MaxSalary']] = data_analyst_jobs['salary_estimate'].str.extract(r'\$(\d+)K-\$(\d+)K')
```

```python
data_analyst_jobs[['MinSalary', 'MaxSalary']]
```

```python
# 8. Finding the datatype of Min and Max Salary
data_analyst_jobs['MinSalary'].dtype, data_analyst_jobs['MaxSalary'].dtype
```

```python
# 9. Converting the Min and Max Salary into float values
data_analyst_jobs['MinSalary'] = data_analyst_jobs['MinSalary'].astype(float)
data_analyst_jobs['MaxSalary'] = data_analyst_jobs['MaxSalary'].astype(float)
```

```python
# 10. To find the average salary
data_analyst_jobs['average_salary'] = (data_analyst_jobs['MaxSalary'] + data_analyst_jobs['MinSalary']) / 2
```

```python
# Removing the columns with salary estimates, min and max salary estimates
data_analyst_jobs.drop(['salary_estimate', 'MinSalary', 'MaxSalary'], axis = 1, inplace = True)
```

The Data Analyst Job have been analysed in four broad categores: - Average Salary, Ratings, Size and Average Revenue.

These factors have further been divided into other factors like job titles, locations, sectors, industries and type of ownerships to gauge the trend in the job market.
